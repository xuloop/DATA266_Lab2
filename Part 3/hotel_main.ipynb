```python
from crewai import Agent, Crew, Task
from hotel_tool import get_hotel_data_combined

# Define the agent
hotel_agent = Agent(
    role="Hotel Travel Assistant",
    goal="Help users find and compare hotel offers with ratings in a given city and date range",
    backstory=(
        "You are an experienced travel assistant specializing in hotel booking and comparison. "
        "You can search hotels, check availability, and summarize pricing, location, and guest ratings."
    ),
    tools=[get_hotel_data_combined],
    verbose=True
)

# Define the task
hotel_task = Task(
    description=(
        "Search for hotels in the city code 'LON' from 2025-05-05 to 2025-05-10 for 2 adults. "
        "Display name, location, price, room type, and rating."
    ),
    expected_output="List 3 good hotels and recommend the best one based on ratings, prices, and room descriptions.",
    agent=hotel_agent
)

# Assemble and run the crew
crew = Crew(
    agents=[hotel_agent],
    tasks=[hotel_task],
    verbose=True
)

results = crew.kickoff()
print("\n=== FINAL RESULTS ===\n", results)
```
