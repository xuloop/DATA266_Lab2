```python
from crewai import Agent, Crew, Task
from weather_tool import get_weather_data

# Define the Weather Agent with backstory
weather_agent = Agent(
    role="Weather Assistant",
    goal="Provide weather forecasts in London for next 4 days.",
    tools=[get_weather_data],
    backstory="You are a helpful assistant that provides accurate weather forecasts for cities around the world.",
    verbose=True
)

# Define a Task for the agent
weather_task = Task(
    agent=weather_agent,
    description="Retrieve the weather forecast for next 4 days.",
    expected_output="Select the day has good weather for going out",
)

# Create a Crew and assign the task
crew = Crew(
    agents=[weather_agent],
    tasks=[weather_task],
    verbose=True  # Enable detailed logs for debugging
)

# Run the task
result = crew.kickoff()

# Print the result
print("\n=== FINAL RESULTS ===\n", result)
```
