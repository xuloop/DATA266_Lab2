```python
from crewai import Agent, Crew, Task
from flight_tool import search_flights
from hotel_tool import get_hotel_data_combined
from weather_tool import get_weather_data
from datetime import datetime, timedelta

# Define your all-in-one travel planner agent
planner_agent = Agent(
    role="Travel Planner Assistant",
    goal="Create a full travel plan with flights, hotels, and weather forecast.",
    tools=[search_flights, get_hotel_data_combined, get_weather_data],
    backstory="You're an experienced travel assistant who finds the best travel options and weather insights for any trip.",
    verbose=True
)

# Define the unified travel task
travel_plan_task = Task(
    agent=planner_agent,
    description="""
        Plan a complete trip from SFO to PAR from 2025-05-05 to 2025-05-10 for 2 people.
        Use tools to find:
        1.  Flight options.
        2.  Best hotel options.
        3.  Weather forecasts.
        Present the data in an easy-to-read travel itinerary format.
    """,
    input_parameters=["origin", "destination", "start_date", "end_date"],
    expected_output="A full travel plan itinerary with recommendation for the best 2 flights, hotels, and weather in the well format"
)

# Create the crew with just the planner agent
crew = Crew(
    agents=[planner_agent],
    tasks=[travel_plan_task],
    verbose=True
)

# Kick off the planning process
results = crew.kickoff()

# Display the results
print("\n === COMPLETE TRAVEL PLAN === \n", results)
```
