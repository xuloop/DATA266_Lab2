```python
from crewai import Agent, Crew, Task
#from crew_agents import flight_agent  
from flight_tool import search_flights  

# Register the tool with the agent
#flight_agent.tools = [search_flights]

# Define the agent
flight_agent = Agent(
    role="Flight Search Agent",
    goal="Find the best flight based on user needs",
    backstory="You are a skilled travel assistant specializing in flights.",
    tools=[search_flights],
    verbose=True
)

# Create a flight search task
flight_task = Task(
    description="Find the 2 nonstop flights from NYC to LON for 2 people, departing on 2025-05-05 and returning on 2025-05-10.",
    expected_output="List 3 good flight options and recommend the best one.",
    agent=flight_agent
)

# Create a Crew and assign the task
crew = Crew(
    agents=[flight_agent],
    tasks=[flight_task],
    verbose=True  # Enable detailed logs for debugging
)

# Run the crew
if __name__ == "__main__":
    results = crew.kickoff()
    print("\n=== FINAL RESULTS ===\n", results)
```
